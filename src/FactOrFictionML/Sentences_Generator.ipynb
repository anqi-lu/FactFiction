```python
from nltk.corpus import wordnet as wn
```

```python
sub_templates = [
    'This <Thing> is very <Adjective>.',
    'This is a very <Adjective> <Thing>.'
]
obj_templates = [
    'This is the <Thing>.',
    'That is the <Thing>.',
    'A <Thing> is a <Thing>.',
    'The <Thing> is a <Thing>.'
]
```

```python
tokens = {
    '<Thing>': 'time year people way day man thing woman life child world school state family student group country problem hand part place case week company system program question work government number night point home room mother area money story fact month lot right study book eye job word business issue side kind head house service friend father power hour game line end member law car city community president team minute idea kid body information back parent face others level office door health person art war history party result change morning reason research girl guy moment air teacher force education'.split(),
    '<Adjective>': 'best silly funny stupid awful'.split()
}
```

```python
# objective
k = '<Thing>'
objectives = []
for t in obj_templates:
    objectives.extend([ t.replace(k, w) for w in tokens[k] ])

k = '<Thing>'
temp = []
for t in sub_templates:
    temp.extend([ t.replace(k, w) for w in tokens[k] ])

subjective = []
k = '<Adjective>'
for t in temp:
    subjective.extend([ t.replace(k, w) for w in tokens[k] ])
```

```python
import random
sampled = random.sample(subjective, len(objectives))
for w in sampled:
    print(w)
```
